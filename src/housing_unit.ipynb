```python
import util.data_loading
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
house_df1 = util.data_loading.load_raw_data('housing_units_00-09.csv')
```

```python
house_df1['Area'] = house_df1['Area'].str.replace('.', '')
```

```python
house_df1 = house_df1.drop("April 1 Estimate", axis=1)
house_df1 = house_df1.drop("2011", axis=1)
```

```python
house_df1
```

```python
house_df2 = util.data_loading.load_raw_data('housing_units_10-19.csv')
house_df2['Area'] = house_df2['Area'].str.replace('.', '')
house_df2 = house_df2.drop("April 1 Estimate", axis=1)
house_df2 = house_df2.drop("2020", axis=1)
```

```python
house_df2
```

```python
house_df3 = util.data_loading.load_raw_data('housing_units_20-24.csv')
house_df3['Area'] = house_df3['Area'].str.replace('.', '')
house_df3 = house_df3.drop("April 1 Estimate", axis=1)
house_df3['Area'] = house_df3['Area'].str.replace(', California', '')
```

```python
house_df3
```

```python
house_all = house_df1.merge(house_df2, on='Area', how="outer").merge(house_df3, on='Area', how="outer")
```

```python
house_all
```

### remove commas convert all to int

```python
house_all.iloc[:, 1:] = house_all.iloc[:, 1:].apply(
    lambda col: pd.to_numeric(col.astype(str).str.replace(',', ''))
)
```

```python
house_all
```

```python
cal_df = house_all[house_all["Area"] ==  "California"].iloc[0]

cal_df
```

```python
years = cal_df.index[1:]  # skip the first column if it's Area
values = cal_df.values[1:].astype(float)
```

```python
sns.scatterplot(x=years, y=values)
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
