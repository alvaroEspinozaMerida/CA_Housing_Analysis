```python
import util.data_loading
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = util.data_loading.load_raw_data('CAPOP.csv')
```

```python
df
```

```python
df.shape
```

```python
df = df.rename(columns={'observation_date': 'year', 'CAPOP': 'population'})
```

```python
df
```

```python
df["year"] =  pd.to_datetime(df["year"], format='mixed')
df["year"] = df["year"].dt.year
df["population"] = df["population"].astype(float)
```

```python
# df = df[df["year"] > 1999]
```

```python
df
```

```python
plt.figure(figsize=(14,6))
sns.scatterplot(data = df, x = "year", y ="population")
ax = plt.gca()

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
df.to_csv("population_clean.csv", index=False, header=True, sep=",")
```
