```python
import util.data_loading
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def create_scatter_plot(df, city_name):

    plt.figure(figsize=(14,6))
    sns.scatterplot(data = df, x = "Mon-Yr", y =city_name)


    ax = plt.gca()



    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

def descriptive_stats(df,  city_name):
    stats = df[city_name].describe()
    print(stats)

    print(f"\nMean: ${stats['mean']:.2f}")
    print(f"Median: ${df['CA'].median():.2f}")
    print(f"Standard Deviation: ${stats['std']:.2f}")
    print(f"Min: ${stats['min']:.2f}")
    print(f"Max: ${stats['max']:.2f}")
    print(f"Range: ${stats['max'] - stats['min']:.2f}")
```
